## Teilaufgabe 1.3.3

```python
import matplotlib.pyplot as plt
import numpy as np

anzahl_zimmer = [1, 2, 3, 4, 5]
absolute_h = [3, 1, 3, 1, 2]
relative_h = [0.3, 0.1, 0.3, 0.1, 0.2] 

cumulative_rel_freq = np.cumsum(relative_h)

plt.rc('axes', axisbelow=True)
plt.xticks(range(max(anzahl_zimmer) + 1))
plt.grid(color='gray', linestyle='-', linewidth=1)
plt.xlabel("Anzahl Zimmer")
plt.ylabel("Kumulierte Häufigkeit")
plt.step(anzahl_zimmer, cumulative_rel_freq, where='post')
plt.show()
```

```python
import numpy as np
import statistics as stat

stichprobe = [1, 1, 1, 2, 3, 3, 3, 4, 5, 5];

print("a) Arithmetisches Mittel: ", np.mean(stichprobe))
print("b) Median: ", np.median(stichprobe))
print("c) Modalwert: ", stat.multimode(stichprobe))
print("d) 25% Quantil: ", np.quantile(stichprobe, 0.25))
print("e) 30% Quantil: ", np.quantile(stichprobe, 0.3))
print("f) 66% Quantil: ", np.quantile(stichprobe, 0.66))
print("g) Varianz: ", np.var(stichprobe))
print("h) Standardabweichung: ", np.std(stichprobe, ddof=1))
print("i) Interquartilabstand: ", np.quantile(stichprobe, 0.75) - np.quantile(stichprobe, 0.25))
print("j) Spannweite: ", np.ptp(stichprobe))
```

```python
import numpy as np

ad = np.array([9, 13, 15, 18, 20])
jg = np.array([18, 37, 61, 125, 59])

korrelation = np.corrcoef(ad, jg)[0, 1]
print(f"Korrelationskoeffizient beträgt: ", korrelation)

# als Matrix
# korrelation = np.corrcoef(ad, jg)
# print(f"Korrelationskoeffizient: ", korrelation)
```
