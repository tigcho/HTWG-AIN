### Teilaufgabe 4.7.2

```python
import numpy as np

print('a)\nErwartungswert von X1:')
werte = np.asarray([1, 2, 3])
probs = np.asarray([1/3, 1/3, 1/3])
erwar = (werte*probs).sum()
print(erwar)
print('\nb)\nVarianz von X1:')
var = 0
for i in range(len(werte)):
    var += (werte[i] - erwar)**2*probs[i]
print(var)
print('\nc)\nStandartabweichung von X1:')
print(np.sqrt(var))

print('\nd)\nErwartungswert von X3:')
werte = np.asarray([2, 3, 4, 5, 6])
probs = np.asarray([1/9, 2/9, 3/9, 2/9, 1/9])
erwar = (werte*probs).sum()
print(erwar)
print('\nVarianz von X3:')
var = 0
for i in range(len(werte)):
    var += (werte[i] - erwar)**2*probs[i]
print(var)
```
