    T: Ein Kunde gehört zur Gruppe der Normalos = 50%
    C: Ein Kunde gehört zur Gruppe der Chaoten = 100% − P(Streber) − P(Normalos) = 30%
    V|S: Ein Streber verliert die Kreditkarte = 1.5%
    V|T: Ein Normalo verliert die Kreditkarte = 3%
    V|C: Ein Chaot verliert die Kreditkarte = 5.5%
b)
P(Kreditkartenverlust) <br> = P(Kreditkartenverlust | Streber) × P(Streber) + P(Kreditkartenverlust | Normalos) × P(Normalos) + P(Kreditkartenverlust | Chaoten) × P(Chaoten)
P(Kreditkartenverlust) <br> = 0.015 × 0.2 + 0.03 × 0.5 + 0.055 × 0.3 = 0.0345 = 3.45%

c) 
Streber: P(S|V) = (P(V|S) * P(S)) / P(V) = (0.015 * 0.20) / 0.0345 = 0.08695 ≈ 0.0867 = 8.67% <br>
Normalo: P(T|V) = (P(T|S) * P(T)) / P(V) = (0.03 ⋅ 0.50) / 0.0345 = 0.43478 ≈ 0.435 = 43.5% <br>
Chaot: P(C|V) = (P(C|S) * P(C)) / P(V) = (0.055 ⋅ 0.30) / 0.0345 = 0.47826 ≈ 0.478 = 47.8% <br>
